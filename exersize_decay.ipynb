```python
#فراخوانی کتابخانه ها
import matplotlib.pyplot as plt
import math
import numpy as np
import sympy as sp
from math import pi
```

Simulation of the particle decay
dNA/dt=(NB/tau)-(NA/tau)
dNB/dt=(NA/tau)-(NB/tau)
NA(t=0)=100
NB(t=0)=0


```python
#dNA/dt=f(t,NA)
def fA(t,NA,NB):
   return (NB/tau)-(NA/tau)


#dNB/dt=f(t,NB)
def fB(t,NA,NB):
   return (NA/tau)-(NB/tau)
```

```python
#get the inital data
#t1=inital time 
t1=float(input("Enter the initial time"))
tfinal=float(input("Enter the final time:"))
#N1=inital NA,NB
NA1=float(input("Enter the initial value of NA :"))
NB1=float(input("Enter the initial value of NB :"))
#tau=time constant
tau=float(input("Enter the time constant(tau):"))
h=float(input("Enter the h :"))
#draw diagram
fig0=plt.figure()

ax=plt.subplot()
plt.xlabel('time(s)')
plt.ylabel('NA(t),NB(t)')
plt.title("numerical and analytical solution of nuclear decay NA,NB")  
n=int((tfinal-t1)/h)
NA0=NA1
NB0=NB1
t0=t1
for i in range(0,n):
 
    
    sA1=fA(t1,NA1,NB1)
    sB1=fB(t1,NA1,NB1)
    
    t2=t1+h
    
    NB2=NB1+h*sB1
    NA2=NA1+h*sA1

    sA2=fA(t2,NA2,NB2)
    sB2=fB(t2,NA2,NB2)
    
    NA2=NA1+h*(sA1+sA2)/2
    
    
    NB2=NB1+h*(sB1+sB2)/2
   
    ax.scatter(t1,NA1,color='red',marker='o',s=5)
    ax.scatter(t1,NB1,color='blue',marker='o',s=5)    
    t1=t2
    NA1=NA2
    
    NB1=NB2
    i=i+1



#Analytical solution
#if we have tau=1,by analytically soving the above differential equation we obtian the following results:
#t=time
t=np.linspace(t0,tfinal,100)
NAT=(NA0+NB0)*(1+np.exp((-2*t)/tau))/2
NBT=(NA0+NB0)*(1-np.exp((-2*t)/tau))/2

plt.plot(t,NAT,color='black',linestyle='dashed',label='Analytical solution')
plt.plot(t,NBT,color='black',linestyle='dashed')                                      
plt.legend()
plt.savefig('decay_exersize')
plt.grid(True)         

    

```
