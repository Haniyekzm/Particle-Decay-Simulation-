```python
 #فراخوانی کتابخانه ها
import matplotlib.pyplot as plt
import math
import numpy as np
import sympy as sp

#this code is for particle decay
#the equation of that is dN/dt=-N/tau ;N(t)=N(0)exp(-t/tau)
#get the initial data
t1=float(input("Enter the initial value of time :"))
tfinal=float(input("Enter the final time :"))
tau=float(input("Enter the time constant for decay :"))
y1=float(input("Enter the N(0) :"))
h=float(input("Enter the h :"))


#dN/dt=f
#y1=N(0)
#y=N
#N(t)=N(0)exp(-t/tau)
#N(0)=y1

def f(t,y):
    return -y1*(math.exp(-t/tau))
    
y=np.linspace(0,100,1000)

#draw the figure

fig =plt.figure()
plt.xlabel('time')
plt.ylabel("N(t)")
ax=plt.subplot()
i=0
while t1<=tfinal:
    y2=y1+h*f(t1,y1)
    t2=t1+h
    y1=y2
    t1=t2
    ax.scatter(t1,y1)
    i+=1
    if i>5000:
        break
print(" number of remaining particles :",y1)
plt.show()
```
